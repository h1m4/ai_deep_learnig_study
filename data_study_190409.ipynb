# 5.4.2 데이터 누락 처리하기
 * dropna
 
  - df.dropna() : 하나 이상의 열에 누락된 값이 있는 경우 전부 삭제
  - df.dropna(how='all') : 해당 행의 모든 값이 누락된 경우 삭제
  - df.dropna(axis=1, thresh=3) : row에서 NaN이 아닌 값이 최소 3개 이상 나오면 남김

```
df.dropna()
```

# 데이터 누락 값 채우기
 - df.fillna(0) : 모든 누락 값을 같은 값으로 채우기, 이 경우 '0'
 - df.fillna(method='pad') : 누락값을 그 이전 값으로 채우기
 

```
df.fillna(method='pad') # 그 이전 열의 값으로 채움
```

```
df.fillna(method='bfill') # 그 다음 열 값으로 채움
```

```
df.fillna({1: 200, 3: 100}) # 그 열에 있는 숫자만 같은 숫자로 채우기
```

```
import numpy as np
df2 = pd.DataFrame(np.arange(1, 26).reshape(5, 5))
df2
```

```
df.fillna(df2) # 맞는 숫자에 넣기
```

```
_ = df.fillna(0, inplace = True) # 0을... 채운다...?
df
```

```
df.fillna(df.mean()) # 열의 중간값 채우기
```

# 계층 색인
* 행이나 열에 여러 계층을 가지는 인덱스를 만들 수 있다. 데이터프레임을 생성할 때 index에 2차원 배열로 형태로 인덱스를 넣으면 다중 열 인덱스를 가질 수 있음.
* 밑 예시는 index만, colums로 하면 columns 생김

## 열색인

```
data = Series(np.random.randn(10), index = [['a', 'a', 'a', 'b', 'b', 'b', 'c', 'c', 'd', 'd'], [1, 2, 3, 1, 2, 3, 1, 2, 2, 3]])
data
```

```
data.index
```

```
data['b'] # 데이터에 대한 부분적 색인 접근
```

```
data[:, 2]
```

```
data.unstack() # 데이터를 재형성하고 피벗 테이블 생성
```

```
data.unstack().stack() # 7장에서 자세히 한다구 함, stack() : 열 인덱스 -> 행 인덱스로 변환
```

## 행색인

```
frame = DataFrame(np.arange(12).reshape((4, 3)), index = [['a', 'a', 'b', 'b'], [1, 1, 2, 2]], columns = [['Ohio', 'Ohio','Colorado'], ['Green', 'Red', 'Green']])
frame
```

### 이름 붙여주기

```
frame.index.names = ['key1', 'key2']
frame.columns.names = ['state', 'color']
frame
```

## 5.5.1 계층 순서 바꾸고 정렬하기

- swaplevel
- sortlevel

```
frame.swaplevel('key1', 'key2')
```

```
frame.sortlevel(1) #뭐임... : 단일 계층에 속한 데이터를 정렬한다.
```

```
frame.swaplevel(0, 1).sortlevel(0) #머임..
```

## 5.5.2 단계별 요약통계

```
frame.sum(level = 'key1')
```

```
frame.sum(level = 'color', axis = 1)
```
